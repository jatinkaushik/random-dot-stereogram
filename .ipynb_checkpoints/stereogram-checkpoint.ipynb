# Random Dot Stereogram

Python code for generating a red blue stereogram with two rectangles popping out of the middle.

### Import statements

```python
import numpy as numpy
import matplotlib.pyplot as pyplot
import matplotlib.colors as colors
%matplotlib inline
```

### Constants

```python
# Colors to match red blue glasses
MY_BLUE = colors.hex2color('#00FFFF')
MY_RED = colors.hex2color('#FF0000')

# Plot dimensions
PLOT_WIDTH = 15
PLOT_LENGTH = 10
NUM_DOTS = 5000
RB_DOT_ADJUST = 0.005

# Rectangle dimensions
PLOT_BORDER = 0.2
RECT_WIDTH = 0.2
STEREOGRAM_ADJUST = 0.015
```

### Generate random coordinates

```python
red_x_vals = np.random.rand(NUM_DOTS)
red_y_vals = np.random.rand(NUM_DOTS)

cyan_x_vals = red_x_vals + HORIZONTAL_ADJUST 
cyan_y_vals = red_y_vals - VERTICAL_ADJUST
```

### Visualize dots

```python
plot = pyplot.gcf()
plot.set_size_inches((PLOT_WIDTH, PLOT_LENGTH))
cyan_dots = pyplot.scatter(cyan_x_vals, cyan_y_vals, color=MY_BLUE)
red_dots = pyplot.scatter(red_x_vals, red_y_vals, color=MY_RED)
pyplot.xlim([0,1])
pyplot.ylim([0,1])
pyplot.axis('off')
```

### Create rectangles that 'pop out' of the image

```python
left_rect = (red_x_vals >= PLOT_BORDER) \
    & (red_x_vals <= PLOT_BORDER + RECT_WIDTH) \
    & (red_y_vals >= PLOT_BORDER) \
    & (red_y_vals <= 1 - PLOT_BORDER)

right_rect = (red_x_vals >= 2 * PLOT_BORDER + RECT_WIDTH) \
    & (red_x_vals <= 1 - PLOT_BORDER) \
    & (red_y_vals >= PLOT_BORDER) \
    & (red_y_vals <= 1 - PLOT_BORDER)

red_x_vals[left_rect] += STEREOGRAM_ADJUST
red_y_vals[left_rect] -= STEREOGRAM_ADJUST

red_x_vals[right_rect] += STEREOGRAM_ADJUST
red_y_vals[right_rect] -= STEREOGRAM_ADJUST
```

```python
plot = pyplot.gcf()
plot.set_size_inches((PLOT_WIDTH, PLOT_LENGTH))
pyplot.scatter(cyan_x_vals, cyan_y_vals, color=MY_BLUE)
pyplot.scatter(red_x_vals, red_y_vals, color=MY_RED)
pyplot.xlim([0,1])
pyplot.ylim([0,1])
pyplot.axis('off')
```
